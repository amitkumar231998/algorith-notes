```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

```python
img=cv2.imread('6.jpg')
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

# Rotating the image

```python
imgr=cv2.rotate(img,rotateCode=0)
```

```python
plt.imshow(imgr)
plt.show()
```

#  High pass filter using lapacian method having scale =1

```python
edges=cv2.Laplacian(img,-1,ksize=3,scale=1,borderType=cv2.BORDER_DEFAULT)
```

```python
output=[img,edges]
```

```python
for i in range(2):
    plt.imshow(output[i])
plt.show()
```

#  low pass filter and using blur 

```python
blur=cv2.blur(img,(3,3))
```

```python
outputs_for_low_filter=[img,blur]
```

```python
for j in range(2):
    plt.imshow(outputs_for_low_filter[j])
plt.show()
```

# writing the image

```python
cv2.imwrite('image.jpg',imgr)
```
